```python
import os
import asyncio
from random import randint
from typing import Annotated
from dotenv import load_dotenv

from agent_framework.azure import AzureAIProjectAgentProvider
from azure.identity.aio import AzureCliCredential
from pydantic import Field
```

```python
load_dotenv()
```

```python

async with (
        AzureCliCredential() as credential,
        AzureAIProjectAgentProvider(project_endpoint=os.environ["AZURE_AI_PROJECT_ENDPOINT"], credential=credential) as provider,
    ):
        agent = await provider.create_agent(
            name="Agent-Framework-Haiku-Writer",
            instructions="Write a haiku about Agent Framework",
        )

        query = "Write a haiku about Agent Framework"
        print(f"User: {query}")
        print("Agent: ", end="", flush=True)
        async for chunk in agent.run_stream(query):
            if chunk.text:
                print(chunk.text, end="", flush=True)
        print("\n")
```
